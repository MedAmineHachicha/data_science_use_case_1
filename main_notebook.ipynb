### Categorical Variables
From the variable name, this variable propably tells the number of kids that can drive the vehicle.

```python
categ_variables = ['KIDSDRIV', 'HOMEKIDS', 'PARENT1', 'MSTATUS', 'SEX', 'EDUCATION', 'JOB', 'CAR_USE', 
                   'CAR_TYPE', 'RED_CAR', 'REVOKED', 'URBANICITY']

for categ_var in categ_variables:
    plot_categorical(data_train, col=categ_var, target="TARGET_FLAG")
```

#### We take the example of KIDSDRIV: 

We can see from the plots above that:

* The number of kids that drive is decreasing, with most clients having 0 kids that drive the vehicle.
* TARGET_FLAG has more 1 labelled observations when KIDSDRIV increases. Therefore, the two variables are not independent and we will keep KIDSDRIV

We will treat this variable as a continuous variable when fitting our models, since the categories are ordered and we see that KIDSDRIV and TARGET_FLAG are correlated.

We will also treat HOMEKIDS as a continuous variables

#### In general:
We can see that for most variables (except for SEX and RED_CAR)
    * The categories populations are unbalanced
    * The proportions of TARGET_FLAG depend on the category, which means that TARGET_FLAG is not independent from these categorical variables. We can see for example with the variable CAR_USE that cars for commercial use are more prone to have accidents. This is probably because commercial cars are driven for longer distances than private cars.
    We can also see for example that cars in Urban areas are more prone to have accidents. This is logical since traffic is much higher in Urban areas than in Rural areas.
    
#### For variables SEX and RED_CAR:
    * We can see for these variables that the average number of accidents doesn't depend on the category.
    * We do a Chi2 test to determine statistically if these variables are relevant.

```python
vars_to_test = ["SEX", "RED_CAR"]

for var in vars_to_test:
    p_value = get_chi2_pval(df=data_train, col=var, target="TARGET_FLAG")
    print(f"{var} p-value for dependeny test with TARGET_FLAG: {p_value}")
```

$\rightarrow$ In a Chi-Squared test the null hypothesis is that the two variables are independent. A small p-value (<0.05) means that we can reject this hypothesis.
    
    Both tests give a high p-value (>0,05) which means that there's no proof of dependency between TARGET_FLAG and the two variables SEX and RED_CAR. Therefore we will eliminate these two variables.

#### N.B.: In real life, many country regulations prohibit using some variables such as SEX or RELIGION to calculate insurance primes. Here SEX is statistically irrelevant, but even if it adds information, we may have to delete it in real life applications

## Feature Engineering

The HOME_VAL and INCOME are two variables that should be considered with regards to the houshold size. A household that has an income of 60k dollars but with 5 members is less financially well-off than a houshold of 2 members with an income of 50k dollars. Therefore we will add two variables: HOME_VAL_PER_CAPITA and INCOME_PER_CAPITA where we respectively divide HOME_VAL and INCOME by the household size.

```python
data_train['HOME_VAL_PER_CAPITA'] = data_train['HOME_VAL'] / (1 + (data_train['MSTATUS']=='Yes') + data_train['HOMEKIDS'])
data_test['HOME_VAL_PER_CAPITA'] = data_test['HOME_VAL'] / (1 + (data_test['MSTATUS']=='Yes') + data_test['HOMEKIDS'])

data_train['INCOME_PER_CAPITA'] = data_train['INCOME'] / (1 + (data_train['MSTATUS']=='Yes') + data_train['HOMEKIDS'])
data_test['INCOME_PER_CAPITA'] = data_test['INCOME'] / (1 + (data_test['MSTATUS']=='Yes') + data_test['HOMEKIDS'])
```

Also, as we explained above we will split the variable JOB into two categories : MANAGER_DOCTOR and OTHERS

```python
data_train.loc[data_train["JOB"].isin(["Doctor", "Manager"]), "JOB"] = "MANAGER_DOCTOR"
data_train.loc[(data_train["JOB"]!="MANAGER_DOCTOR") & (~data_train["JOB"].isna()), "JOB"] = "OTHERS"

data_test.loc[data_test["JOB"].isin(["Doctor", "Manager"]), "JOB"] = "MANAGER_DOCTOR"
data_test.loc[(data_test["JOB"]!="MANAGER_DOCTOR") & (~data_test["JOB"].isna()), "JOB"] = "OTHERS"
```

## Data Preparation

#### For categorical variables we replace missing values with the most frequent category. Then we do a OneHotEncoding
#### For continuous variables, we replace missing values with the mean and standardize the variables

```python
from feature_processing import FeatureProcessor
FP = FeatureProcessor(categ_cols=categ_variables, cont_cols=cont_variables+["HOME_VAL_PER_CAPITA", "INCOME_PER_CAPITA"])
FP.preprocessor.fit(data_train)

X_train = FP.transform(data_train)
X_test = FP.transform(data_test)

y_train = data_train['TARGET_FLAG']
```

## Models Tuning with GridSearch and Cross-Validation

```python
# Split train set into train set and dev set

from sklearn.model_selection import train_test_split
X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train, test_size=0.2, shuffle=True, random_state=123)
```

```python
from grids import RF_grid, SVM_grid, XGB_grid
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import balanced_accuracy_score, make_scorer
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC

scorer = make_scorer(balanced_accuracy_score)
```

#### RandomForest

```python
rf = RandomForestClassifier()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions=RF_grid, n_iter = 50, 
                               cv = 3, verbose=1, random_state=123, n_jobs = -1, scoring=scorer)

rf_random.fit(X_train, y_train)
```

#### SVM Classifier

```python
svc_clf = SVC()
svc_random = RandomizedSearchCV(estimator = svc_clf, param_distributions=SVM_grid, n_iter = 50, 
                               cv = 3, verbose=1, random_state=123, n_jobs = -1, scoring=scorer)

svc_random.fit(X_train, y_train)
```

#### XGBoost

```python


xgb = XGBClassifier(learning_rate=0.01, n_estimators=200, objective='binary:logistic',
                    nthread=3, use_label_encoder=False, eval_metric="logloss")
xgb_random = RandomizedSearchCV(xgb, param_distributions=XGB_grid, n_iter=30, scoring=scorer, 
                                   n_jobs=-1, cv=3, verbose=2, random_state=123 )

xgb_random.fit(X_train, y_train)
```

## Final Model Performance

#### We select the model that has the highest balanced accuracy on the dev set.

```python
models = {"Random Forest": rf_random,
         "SVC": svc_random,
         "XGB": xgb_random}
```

```python
dev_balanced_accs = {}
for model_name in models.keys():
    model = models[model_name]
    y_pred_model = model.predict(X_dev)
    balanced_acc = balanced_accuracy_score(y_dev, y_pred_model)
    dev_balanced_accs[model_name] = balanced_acc
    
print('Balanced Accuracies: ')
print(pd.Series(dev_balanced_accs))
```

```python
from sklearn.metrics import classification_report
best_model_name = max(dev_balanced_accs, key=dev_balanced_accs.get)
print(f"The best model is: {best_model_name}")
print(classification_report(y_dev, models[best_model_name].predict(X_dev)))
```

## Export Predictions

```python
best_model = models[best_model_name]

y_pred = best_model.predict(X_test)

pd.DataFrame(y_pred, index=data_test.INDEX, columns=["TARGET_FLAG"]).to_csv("output/TARGET_FLAG.csv")
```
